### Machine Learning 

#### Baseline; Using Logistics regression on all variables

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
x = data.drop("target", axis=1).values
y = data.target.values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
```

```python
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy_score(y_test, y_pred)
```

```python
rf = RandomForestClassifier(max_depth = 8, random_state = 42, n_estimators= 1000)
rf.fit(x_train, y_train)

# make predictions
y_pred = rf.predict(x_test)

# evaluate model
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gra = GradientBoostingClassifier()
gra.fit(x_train, y_train)


y_pred = gra.predict(x_test)

accuracy_score(y_test, y_pred)
```

### Droping of some coloumns 
Resting ECG has a weak correlation with the occurrence of heart disease, will therefore be removed
Cholesterol also has a negative weak correlation and will also be dropped from the dataset
Alongside fasting blood sugar and trestbps

```python
df=data.drop(["trestbps", "chol", "fbs", "restecg"], axis=1)
df
```

```python
X= df.drop("target", axis=1).values
Y=df.target.values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler  ##to improve data scalability

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

accuracy_score(Y_test, Y_pred)
```

#### Random Forest Classifier

```python
rf = RandomForestClassifier(max_depth = 8, random_state = 42, n_estimators= 1000)
rf.fit(X_train, Y_train)

# make predictions
Y_pred = rf.predict(X_test)

# evaluate model
accuracy_score(Y_test, Y_pred)
```

#### GradientBoostingClassifier

```python
from sklearn.ensemble import GradientBoostingClassifier
gra = GradientBoostingClassifier()
gra.fit(X_train, Y_train)


Y_pred = gra.predict(X_test)

accuracy_score(Y_test, Y_pred)
```

### Cross Validation

```python
#Evaluating models
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from numpy import mean
from numpy import std
```

```python
# Define models
models = {
    "Logistic Regression": model,
    "Random Forest": rf,
    "Gradient Boosting": gra
}
```

```python
# Perform 5-Fold Cross Validation for each model
for name, model in models.items():
    scores = cross_val_score(model, X_train, Y_train, cv=5, scoring='accuracy')  # 5-fold cross-validation
    print(f"{name}: Mean Accuracy = {scores.mean():.4f}, Std Dev = {scores.std():.4f}")
```
