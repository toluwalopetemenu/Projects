**Introduction:**
Happiness and wellbeing are of importance to the average person, which can be influenced by various factors, environmental factors are an important factor influencing the level of happiness people feel daily. The country we live in, the conditions available in such countries ultimately also impacts.
From our dataset, the happiness level of 136 countries has been rated between 1-10, with 1 being the happiest and 10 as most happy countries using 6 factors (GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, perceptions of corruption). 

**Body:**
Analysis of countries showed that some countries have higher happiness scores than others 10 countries with the highest happiness scores are: (Finland, Denmark, Iceland, Isreal, Netherlands, Sweden, Norway, Switzerland, Luxembourg and New Zealand) while people living within countries like Afghanistan, Lebanon, Sierra Leone, Zimbabwe, Congo (Kinshasa), Botswana, Malawi, Comoros, Tanzania, Zambia have the lowest happiness score rating. 
Further analysis shows that within regions North America and ANZ has the highest happiness score average of 7.0, this is also followed closely by the Western Europe region with an average of 6.89.
Several factors positively Influencing happiness across the countries includes GDP (Gross Domestic Product) per capita- which measures the total output created through the production of goods and services in a country over a period, the level of social support received by people living in a country, healthy life expectancy and the freedom to make life choices in the country all positively influences the level of happiness. As these factors positively increases, happiness score increases, and a negative change in these factors will also negatively affect
While these factors have considerable impact, a key driver is the level of social support received by citizens, from our data set a correlation of 0.83, followed by GDP per capita 0.78, life expectancy 0.75, and freedom to make life choices 0.66.
Generosity had no correlation with the rate of happiness (0.04) while perception of corruption (0.47) has a moderate correlation with happiness.

**Conclusion and Recommendations**
Countries with high happiness score should continue to maintain and improve all factors, and I would recommend countries to continue to improve the level of social support put in place to ensure continue wellbeing for it nation.

```python
df2=pd.read_csv(r'C:\Users\EliteBook\Downloads\iot_telemetry_data.csv')
df2
```

```python
from datetime import datetime
```

Copied preprocessing

```python
#Function to convert epoch timestamp to Datetime
def convertToDateTime(ts):
    dateTimeObj = datetime.fromtimestamp(ts)
    return dateTimeObj
```

```python
#Convert the epoch timestamp to Datetime and add as a new column to the dataframe
df2['DateTime'] = df2['ts'].apply(lambda x: convertToDateTime(x))
```

```python
#Drop the original epoch timestamp ('ts')
df2 = df2.drop('ts', axis =1 )
```

```python
#Set the DateTime column as the index for time-series analysis
df2.set_index('DateTime', inplace=True)
```

```python
#Replace device ID to a more relatable name
df2['device'] = df2['device'].replace({'b8:27:eb:bf:9d:51': 'Device 1',
                                     '00:0f:00:70:91:0a': 'Device 2',
                                     '1c:bf:ce:15:ec:4d': 'Device 3'})
```

```python
df2
```

### 1. What is the average temperature and humidity across all devices?


```python
fig = plt.figure(figsize=(10, 8))

plt.subplot(1, 2, 1) #indicates this plot is a part of a 1-row, 2-column figure and the 1st figure
averagetemp = df2.groupby('device')['temp'].mean()
averagetemp.plot(kind='barh', title='Average Temperature across Devices')

plt.subplot(1, 2, 2) #indicates this plot is a part of a 1-row, 2-column figure and the 2nd figure
averagehum = df2.groupby('device')['humidity'].mean()
averagehum.plot(kind='barh', title='Average Humidity across Devices')
```
